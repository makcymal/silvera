```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import cm
from matplotlib.colors import LightSource
```

### Griewank function

$f(x,y) = \cfrac{x^2 + y^2}{A} - \cos{x} * \cos{\cfrac{y}{\sqrt2}}$

$\argmin f(x,y) = (0, 0)$

https://www.sfu.ca/~ssurjano/griewank.html

```python
DOMLIM = 20
PLOT_POINTS = 512

A = 4000

x = np.linspace(-DOMLIM, DOMLIM, PLOT_POINTS)
y = np.linspace(-DOMLIM, DOMLIM, PLOT_POINTS)
x, y = np.meshgrid(x, y)
z = (x ** 2 + y ** 2) / A - np.cos(x) * np.cos(y / np.sqrt(2))

fig, ax = plt.subplots(subplot_kw=dict(projection='3d'))

ls = LightSource(90, 1000)
rgb = ls.shade(z, cmap=cm.viridis, vert_exag=0.1, blend_mode='soft')
surf = ax.plot_surface(x, y, z, facecolors=rgb, antialiased=True, shade=True)
```
