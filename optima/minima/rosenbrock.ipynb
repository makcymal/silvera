```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import cm
from matplotlib.colors import LightSource
```

### Rosenbrock function

$f(x,y) = A (y - x^2)^2 + (x - 1)^2$

$\argmin f(x,y) = (1, 1)$

https://www.sfu.ca/~ssurjano/rosen.html

```python
DOMLIM = 10
PLOT_POINTS = 128

A = 100

x = np.linspace(-DOMLIM, DOMLIM, PLOT_POINTS)
y = np.linspace(-DOMLIM, DOMLIM, PLOT_POINTS)
x, y = np.meshgrid(x, y)
z = A * (y - x ** 2) ** 2 + (x - 1) ** 2

fig, ax = plt.subplots(subplot_kw=dict(projection='3d'))

ls = LightSource(90, 1000)
rgb = ls.shade(z, cmap=cm.magma, vert_exag=0.1, blend_mode='soft')
surf = ax.plot_surface(x, y, z, facecolors=rgb, antialiased=True, shade=True)
```
