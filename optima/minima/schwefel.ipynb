### Schwefel function

$f(x,y) = -x \sin{\sqrt{|x|}} - y \sin{\sqrt{|y|}}$

$\argmin f(x,y) = (420.9687, 420.9687)$

https://www.sfu.ca/~ssurjano/schwef.html

```python
schwefel = Func2d(
  lambda x, y: - (x * np.sin(np.sqrt(np.abs(x))) + y * np.sin(np.sqrt(np.abs(y)))),
  domain_lim=500, argmin=(420.9687, 420.9687) 
)
schwefel.plot(512, 'schwefel')
```
