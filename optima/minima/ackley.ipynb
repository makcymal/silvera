```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import cm
from matplotlib.colors import LightSource
```

### Ackley function

$f(x,y) = -A \exp{\left( -B \sqrt{\cfrac{x^2 + y^2}{2}} \right)} - \exp{\left( \cfrac{\cos{Cx} + \cos{Cy}}{2} \right)}$

$\argmin f(x,y) = (0, 0)$

https://www.sfu.ca/~ssurjano/ackley.html

```python
DOMLIM = 40
PLOT_POINTS = 128

A = 20
B = 0.2
C = 2 * np.pi

x = np.linspace(-DOMLIM, DOMLIM, PLOT_POINTS)
y = np.linspace(-DOMLIM, DOMLIM, PLOT_POINTS)
x, y = np.meshgrid(x, y)
z = -A * np.exp(-B * np.sqrt((x ** 2 + y ** 2) / 2)) - np.exp((np.cos(C * x) + np.cos(C * y)) / 2)

fig, ax = plt.subplots(subplot_kw=dict(projection='3d'))

ls = LightSource(90, 1000)
rgb = ls.shade(z, cmap=cm.viridis, vert_exag=0.1, blend_mode='soft')
surf = ax.plot_surface(x, y, z, facecolors=rgb, antialiased=True, shade=True)
```
